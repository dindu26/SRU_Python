```python
consumer_stts_list = ["Residential", "Commercial", "Industrial"]
consumer_status = input("What is your statues? Residential/Commercial/Industrial: ")
if consumer_status in consumer_stts_list :
  print("Valid Consumer.")
  rqrd_units = input("How many units do you want?: ")
  if consumer_status == "Residential":
    bill = int(rqrd_units) * 2
  elif consumer_status == "Commercial":
    bill = int(rqrd_units) * 5
  else :
    bill = int(rqrd_units) * 10
  print("Your bill is:", bill)
else :
  print("Invalid Consumer. Please write correctly whether you is your status? Residential/Commercial/Industrial")
```

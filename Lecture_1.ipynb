```python
num1 = input("Enter the first number:")
num2 = input("Enter the first number:")
add = int(num1) + int(num2)
sub = int(num1) - int(num2)
mult = int(num1) * int(num2)
div = int(num1) / int(num2)
flr = int(num1) // int(num2)
mod = int(num1) % int(num2)
expn = int(num1) ** int(num2)
print("The sum of the two numbers is:", add)
print("The substraction of the two numbers is:", sub)
print("The multiplication of the two numbers is:", mult)
print("The devision of the two numbers is:", div)
print("The floor of this two numbers:", flr)
print("The mod of two numbers:", mod)
print("The exp of two numbers:", expn)
```
